##### Q1 : 2877. Create a DataFrame from List

#### Q5. 2881. Create a New Column

- A company plans to provide its employees with a bonus.

- Write a solution to create a new column name bonus that contains the doubled values of the salary column.

```python
import pandas as pd

def createBonusColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees_df['bonus'] = employees_df["salary"].apply(lambda x: 2 * x)
    return employees_df

# Test cases:
name = ['Piper', 'Grace', 'Georgia', 'Willow', 'Finn', 'Thomas']
salary = [4548, 28150, 1103, 6593, 74576, 24433]
employees_list = list(zip(name, salary))

employees_df = pd.DataFrame(employees_list, columns=['name', 'salary'])

createBonusColumn(employees_df)
```

#### Q7. 2883. Drop Missing Data

- There are some rows having missing values in the name column.

- Write a solution to remove the rows with missing values.

```python
import pandas as pd

def dropMissingData(students: pd.DataFrame) -> pd.DataFrame:
    students.dropna(subset=['name'], inplace=True)
    return students

# Test case:
student_id = [32, 227, 779, 849]
name = ['Piper', None, 'Georgia', 'Willow']
age = [5, 19, 20, 14]

s_ls = list(zip(student_id, name, age))
s_df = pd.DataFrame(s_ls, columns=['student_id', 'name', 'age'])

dropMissingData(s_df)
```

#### Q8. 2884. Modify Columns

- A company intends to give its employees a pay rise.

- Write a solution to modify the salary column by multiplying each salary by 2.

```python
import pandas as pd

def modifySalaryColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees['salary'] = employees['salary'].apply(lambda x: x*2)
    return employees

#Test case:
name = ['Jack', 'Piper', 'Mia', 'Ulysses']
salary = [19666, 74754, 62509, 54866]
e_ls = list(zip(name, salary))
e_df = pd.DataFrame(e_ls, columns=['name', 'salary'])

modifySalaryColumn(e_df)
```

#### Q9. 2885. Rename Columns

- Write a solution to rename the columns as follows:

-   id to student_id
-   first to first_name
-   last to last_name
-   age to age_in_years

```python
import pandas as pd

def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    students.rename(columns={'id': 'student_id', 'first': 'first_name', 'last': 'last_name', 'age': 'age_in_years'}, inplace=True)
    return students

# Test case:
id = range(1, 6)
first = ['Mason', 'Ava', 'Taylor', 'Georgia', 'Thomas']
last = ['King', 'Wright', 'Hall', 'Thompson', 'Moore']
age = [6, 7, 16, 18, 10]

s_ls = list(zip(id, first, last, age))
s_df = pd.DataFrame(s_ls, columns=['id', 'first', 'last', 'age'])

renameColumns(s_df)
```

#### Q11. 2887. Fill Missing Data

- Write a solution to fill in the missing value as 0 in the quantity column.

```python
import pandas as pd

def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    products.fillna({'quantity': 0}, inplace=True)
    products['quantity'] = products['quantity'].astype(int)
    return products

# Test case:
name = ['Wristwatch', 'WirelessEarbuds', 'GolfClubs', 'Printer']
quantity = [None, None, 779, 849]
price = [135, 821, None, 3051]

p_ls = list(zip(name, quantity, price))
p_df = pd.DataFrame(p_ls, columns=['name', 'quantity', 'price'])

fillMissingValues(p_df)
```

#### Q12. 2888. Reshape Data: Concatenate

- Write a solution to concatenate these two DataFrames vertically into one DataFrame.

```python
import pandas as pd

def concatenateTables(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    df1 = pd.concat([df1, df2])
    return df1

# Test case:
student_id = range(1, 5)
name = ['Mason', 'Ava', 'Taylo', 'Georgia']
age = [8, 6, 15, 17]
df1 = pd.DataFrame(list(zip(student_id, name, age)), columns=['student_id', 'name', 'age'])

student_id = [5, 6]
name = ['Leo', 'Alex']
age = [7, 7]
df2 = pd.DataFrame(list(zip(student_id, name, age)), columns=['student_id', 'name', 'age'])

concatenateTables(df1, df2)
```

#### Q13. 2889. Reshape Data: Pivot

- Write a solution to pivot the data so that each row represents temperatures for a specific month, and each city is a separate column.

```python
import pandas as pd
import numpy as np

def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
    return weather.pivot(columns='city', index='month', values='temp')

# Test case:
city = ['Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville', 'ElPaso', 'ElPaso', 'ElPaso', 'ElPaso', 'ElPaso', ]
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jan', 'Feb', 'Mar', 'Apr', 'May']
np.random.seed(100)
temp = np.random.randint(1, 100, 10)
w_df = pd.DataFrame(list(zip(city, month, temp)), columns=['city', 'month', 'temp'])

pivotTable(w_df)
```

#### Q14. 2890. Reshape Data: Melt

- Write a solution to reshape the data so that each row represents sales data for a product in a specific quarter.

```python
import pandas as pd

def meltTable(report: pd.DataFrame) -> pd.DataFrame:
    unpivot=pd.melt(report,id_vars='product',value_vars=['quarter_1','quarter_2','quarter_3','quarter_4'],var_name='quarter',value_name='sales')
    return unpivot

# Test case:
product = ['Umbrella', 'SleepingBag']
quarter_1 = [417, 800]
quarter_2 = [224, 936]
quarter_3 = [379, 93]
quarter_4 = [611, 875]
df = pd.DataFrame(list(zip(product, quarter_1, quarter_2, quarter_3, quarter_4)), columns=['product', 'quarter_1', 'quarter_2', 'quarter_3', 'quarter_4'])

meltTable(df)
```

#### Q15. 2891. Method Chaining

- Write a solution to list the names of animals that weigh strictly more than 100 kilograms.

- Return the animals sorted by weight in descending order.

```python
import pandas as pd

def findHeavyAnimals(animals: pd.DataFrame) -> pd.DataFrame:
    return animals.loc[animals['weight'] > 100, :].sort_values(by='weight', ascending=False)[['name']]

# Test case:
name = ['Tatiana', 'Khaled', 'Alex', 'Jonathan', 'Stefan', 'Tommy']
species = ['Snake', 'Giraffe', 'Leopard', 'Monkey', 'Bear', 'Panda']
age = [98, 50, 6, 45, 100, 26]
weight = [464, 41, 328, 463, 50, 349]

animals = pd.DataFrame(list(zip(name, species, age, weight)), columns=['name', 'species', 'age', 'weight'])

findHeavyAnimals(animals)
```
